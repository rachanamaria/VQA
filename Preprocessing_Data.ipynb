```python
import json
import shutil
import os



# Load the JSON file
with open('mscoco_train2014_annotations.json', 'r') as f:
    annotations = json.load(f)

# Extract the first 1500 annotations
first_1500_annotations = annotations['annotations'][:1500]

# Save the extracted annotations to a new JSON file
with open('first_1500_annotations.json', 'w') as f:
    json.dump(first_1500_annotations, f, indent=4)

print("First 1500 annotations extracted and saved to 'first_1500_annotations.json' file.")

# Load the JSON annotations
with open('first_1500_annotations.json', 'r') as f:
    annotations = json.load(f)

# Extract image IDs from the annotations
image_ids = set(ann['image_id'] for ann in annotations)

# Specify the path to the MSCOCO 2014 train images directory
train_images_dir = 'train2014'

# Specify the path to the directory where you want to save the extracted images
output_dir = 'extracted_images'

# Create the output directory if it does not exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Copy the images corresponding to the 1500 annotations to the output directory
for image_id in image_ids:
    image_filename = 'COCO_train2014_{:012d}.jpg'.format(image_id)
    image_path = os.path.join(train_images_dir, image_filename)
    output_path = os.path.join(output_dir, image_filename)
    shutil.copy(image_path, output_path)

print("images corresponding to the JSON annotations have been extracted and saved to the '{}' directory.".format(output_dir))


# Load the JSON annotations
# with open('first_1500_annotations.json', 'r') as f:
    # annotations = json.load(f)

# Extract question IDs from the annotations
question_ids = set(ann['question_id'] for ann in annotations)

# Load the questions from the questions_train_mscoco.json file
with open('Questions_Train_mscoco/MultipleChoice_mscoco_train2014_questions.json', 'r') as f:
    questions = json.load(f)

# Extract the corresponding 1500 questions
corresponding_questions = [question for question in questions['questions'] if question['question_id'] in question_ids]

# Save the extracted questions to a new JSON file
with open('corresponding_questions.json', 'w') as f:
    json.dump(corresponding_questions, f, indent=4)

print("Corresponding questions extracted and saved to 'corresponding_questions.json' file.")
```
